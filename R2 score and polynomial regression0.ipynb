```python
import pandas as pd
```

```python
data = [[1,0,0.002],[2,20,0.002],[3,30,0.86]]

df = pd.DataFrame(data,columns=['sn','tem','pres'])
```

```python
df
```

```python
x = df.iloc[:,1:2].values

x
```

```python
y = df.iloc[:,2].values

y
```

```python
from sklearn.linear_model import LinearRegression

regresser = LinearRegression()

regresser.fit(x,y)
```

```python
y_pred = regresser.predict(x)

y_pred
```

```python
import matplotlib.pyplot as plt
```

```python
tem = df['tem'].tolist()
pres = df['pres'].tolist()

plt.plot(tem,pres,color='red')
plt.scatter(tem,pres,color='blue')

plt.xlabel('temparature')
plt.ylabel('pressure')

plt.xticks(tem)
plt.yticks(pres)

plt.title('simple linear regration')

plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=4)

x_poly = poly.fit_transform(x)

x_poly
```

```python
lin = LinearRegression()

lin.fit(x_poly,y)
```

```python
y_poly = lin.predict(x_poly)

y_poly
```

```python
import matplotlib.pyplot as plt

tem = df['tem'].tolist()
pres = df['pres'].tolist()

plt.plot(tem,pres,color='red')
plt.scatter(tem,y_poly,color='blue')

plt.xlabel('temparature')
plt.ylabel('pressure')

plt.xticks(tem)
plt.yticks(pres)

plt.title('simple linear regration')
plt.legend()

plt.show()
```

```python
df = pd.DataFrame({'actual_value':y,'predicted_value':y_pred,'polynomial_value':x_poly})

df
```

```python
from sklearn.metrics import r2_score

print('simple linear regration',r2_score(y,y_pred))
print('polynomial regration',r2_score(y,y_poly))
```
